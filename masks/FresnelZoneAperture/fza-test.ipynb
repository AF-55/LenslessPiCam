```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from sympy.ntheory import isprime, quadratic_residues
#from tqdm import tqdm
```

## Perlin Noise

```python
dim = 300
r = 20
```

```python
def FZA(dim, r):
    x, y = np.meshgrid(np.linspace(-dim/2, dim/2-1, dim), np.linspace(-dim/2, dim/2-1, dim))
    mask = 0.5 * (1 + np.cos(np.pi * (x**2 + y**2) / r**2))
    return mask
```

```python
mask = FZA(dim, r)
plt.imshow(mask, cmap='gray')#, interpolation='lanczos')
plt.colorbar()
```
