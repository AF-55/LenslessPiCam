```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
from math import sqrt
from sympy.ntheory import isprime, quadratic_residues
#from tqdm import tqdm
```

## Fresnel Zone Aperture (FZA)

```python
dim = (510, 255)
r = 30
```

```python
def FZA(dim, r):
    x, y = np.meshgrid(np.linspace(-dim[0]/2, dim[0]/2-1, dim[0]), np.linspace(-dim[1]/2, dim[1]/2-1, dim[1]))
    mask = 0.5 * (1 + np.cos(np.pi * (x**2 + y**2) / r**2))
    return mask
```

```python
mask = FZA(dim, r)
np.save(f"fza_dim{dim[0]}x{dim[1]}_r{r}.npy", mask)
plt.figure(figsize=(13,5))
plt.imshow(mask, cmap='gray')#, interpolation='lanczos')
plt.title(f'Fresnel Zone Aperture (L={dim}, R={r})')
plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.savefig(f'fza_dim{dim[0]}x{dim[1]}_r{r}.png', bbox_inches='tight')
```
